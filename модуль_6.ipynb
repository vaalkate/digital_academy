### Задача 1

Даны четыре действительных числа: x1, y1, x2, y2. Напишите функцию distance(x1, y1, x2, y2), вычисляющая евклидово расстояние ( https://ru.wikipedia.org/wiki/%D0%95%D0%B2%D0%BA%D0%BB%D0%B8%D0%B4%D0%BE%D0%B2%D0%B0_%D0%BC%D0%B5%D1%82%D1%80%D0%B8%D0%BA%D0%B0 ) между точкой (x1,y1) и (x2,y2).

```python
from math import sqrt

x1 = float(input())
y1 = float(input())
x2 = float(input())
y2 = float(input())

def distance(x1, y1, x2, y2):
    return sqrt((x1-x2)**2 + (y1-y2)**2)
print(distance(x1, y1, x2, y2))
```

### Задача 3

Написать функцию is_prime, принимающую 1 аргумент — число от 0 до 1000, и возвращающую True, если оно простое, и False - иначе.

```python
from math import sqrt
print("enter a number from 0 to 1000")
number = int(input())
def is_prime(number):
    for i in range(2, int(sqrt(number)+1)):
        if number % i == 0:
            return False
    return True
print(is_prime(number))
```

### Задача 8

На вход функция more_than_five(lst) получает список из целых чисел. Результатом работы функции должен стать новый список, в котором содержатся только те числа, которые больше 10 по модулю.

```python
lst = [-11, 2, 4, 13, -20, -5]
def more_than_five(lst):
    new_lst = []
    for number in lst:
        if abs(number) > 10:
            new_lst.append(number)
    return new_lst
print(more_than_five(lst))
```

### Задача 13

Написать функцию, принимающую на вход два вектора одинакого размера и находящую косинус угла между векторами. Протестировать на нескольких заданных вами примерах.

```python
import numpy as np
v1 = np.array([1, 2, 3], float)
v2 = np.array([0, 1, 1], float)
v3 = np.array([4, 5, 6], float)
v4 = np.array([7, 8, 1], float)
def cos_of_angle(vector1, vector2):
    numerator = np.dot(vector1, vector2)
    denominator = np.linalg.norm(vector1) * np.linalg.norm(vector2)
    cosin = numerator/denominator
    return cosin
print(cos_of_angle(v1, v2))
print(cos_of_angle(v3, v4))
```

### Задача 16

Написать функцию, принимающую на вход вектор и возвращающую максимальный элемент в векторе среди элементов, перед которыми стоит 0.

Например для:

x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

Ответ: 5

```python
import numpy as np
v1 = np.array([6, 2, 0, 3, 0, 5, 7, 0])
def max_after_zero(vector):
    zero = vector  == 0
    max_number = vector[1:][zero[:-1]].max()
    return max_number
print(max_after_zero(v1))
```

### Задача 17

Написать функцию, принимающую на вход матрицу 5х3 и считающую длину каждого вектора в матрице (строка) и ищущую самый длинный вектор, вернуть его координаты и длину.

Как выглядит матрица:

     | x | y | z |
     | 1 | 2 | 3 |
     | 3 | 4 | 1 |
     | ...       |

```python
import numpy as np
from math import sqrt
arr = np.array([[1, 2, 3],
            [1, 4, 4],
            [2, 3, 3],
            [4, 5, 5],
            [0, 6, 6]])
arr1 = np.random.randint(0, 10, size=(5, 3))
print(arr)
print(arr1)
def longest_vector(array):
    length0 = 0
    for i in range(0, 5):
        x = array[i][0]
        y = array[i][1]
        z = array[i][2]
        print(x,y,z)
        length = sqrt(x ** 2 + y ** 2 + z ** 2)
        print(length)
        if length > length0:
            length0 = length
            coord = array[i]
    return length0, coord
print(longest_vector(arr))
print(longest_vector(arr1))
```
